```python
def sig(x):
    return expit(x)
```

```python
def shuff_data(df):
    water_data_df = df.sample(frac = 1)
    y = water_data_df["Potability"]
    X = water_data_df.drop(["Potability"], axis = 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
    return X_train.to_numpy(), y_train.to_numpy()

    
    
```

```python
def logistic_regression(num_iters, df, eps, lambda_):
    w = np.ones(X_train[0].shape).astype(np.float64)

    for _ in range(num_iters):
        X, y = shuff_data(df)
        for x_i, y_i in zip(X, y):
            w = w + eps*(y_i-sig(w.dot(x_i))*x_i - 2*lambda_*w)
    return w
```

```python
w = logistic_regression(1000, water_data_df, 0.2, 0.9)
```

```python
y_pred = []
for p in sig(X_test.dot(w)):
    if p >= 0.5:
        y_pred.append(1.0)
    else:
        y_pred.append(0.0)
```

```python
log_loss(y_test, y_pred)
#len(np.where(y_test-y_pred == 0)[0])/len(y_test)
```

```python
def kernel_logistic_regression(df, num_iters, alpha, K):
    a = np.ones(K[1].shape[0])
    
    for _ in range(num_iters):
        X, y = shuff_data(df)
        K = X.dot(X.T)
        for y_i, k in zip(y, K):
            a = a + alpha*(y_i-sig(k.dot(a)))
    return a
        
```

```python
a = kernel_logistic_regression(water_data_df, 1000, 0.01, X_train.dot(X_train.T))
```

```python
def classify(a):
    K = X_train.dot(X_test.T)
    y_pred = []

    for p in sig(a.dot(K)):
        if p >= 0.5:
            y_pred.append(1.0)
        else:
            y_pred.append(0.0)
    return y_pred
    
```

```python
def glorot_weights(f_in,f_out):
    avg = (f_in+f_out)/2
    return np.random.normal(0, 1/(np.sqrt(avg)), size=(f_in, f_out))
```

```python
def NNClassifier(X):
    W_1 = glorot_weights(X.shape[0], X.shape[1])
    O_1 = sig(W_1.dot(X.T))
    W_2 = glorot_weights(1, O_1.shape[0])
    O_2 = sig(W_2.dot(O_1.T)).reshape(-1)
    y_pred_nn = []
    
    
    
    
    for p in O_2:
        if p >= 0.5:
            y_pred_nn.append(1.0)
        else:
            y_pred_nn.append(0.0)
    return np.array(y_pred_nn)
        
```
